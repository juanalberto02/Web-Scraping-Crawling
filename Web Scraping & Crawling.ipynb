2.	Lalu, lakukan crawling unair news dengan mengkombinasikan Python Request, beautifulsoup, dan for loop

```python
import requests
import lxml
from bs4 import BeautifulSoup
import pandas as pd
```

import library yang akan kita gunakan

```python
url = "https://unair.ac.id/news"
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}
f = requests.get(url, headers = headers)
```

mirip dengan scraping, kita menggunakan request untuk mendapatkan data html dari link website yang ingin kita tuju 

```python
soup = BeautifulSoup(f.content,'lxml')
```

membuat variabel soup yang berfungsi menjadi wadah dari semua konten yang ada pada website

```python
berita = soup.find_all('h3',{'class':'elementor-post__title'})
```

'.find_all()' berfungsi untuk menemukan semua elemen yang ada pada variabel yang ditentukan

```python
berita
```

```python
linkberita =[]
```

```python
for h3 in berita:
            for link in h3.find_all('a'):
                if link not in linkberita:
                    linkberita.append(link.get('href'))
                else:
                    continue
```

```python
linkberita
```

```python
Kumpulanjudul = []
teks =[]
for i in linkberita:
    url1 = i
    html1 = requests.get(url1)
    soup1 = BeautifulSoup(html1.text, 'html.parser')
    judulberita = soup1.find('h3', {'class': 'elementor-heading-title elementor-size-default'}).text
    text = soup1.find('div', {'class': 'elementor-element elementor-element-e093e29 elementor-widget elementor-widget-theme-post-content'})
    try:
        Kumpulanjudul.append(judulberita)
        teks.append(text.get_text())
        print(judulberita)
        print(text.get_text())
    except AttributeError:
        Kumpulanjudul.pop()
        continue
```

Disini kita menggunakan for loop untuk mendapatkan data dari setiap link yang telah kita peroleh dari website utama (url). cara kerjanya sama seperti scraping pada no 1, namun kita lakukan secara iteratif sampai semua link terpenuhi.

```python
teksbaru = []
for i in teks:
    a = i.replace('\n',' ')
    print(a)
    teksbaru.append(a)
```

Karena terdapat data string yang tidak kita inginkan, kita lakukan sedikit modifikasi agar data lebih mudah untuk dibaca oleh pengguna

```python
s1=pd.Series(Kumpulanjudul,name='Judul')
s2=pd.Series(teksbaru,name='Teks')
crawling_list = pd.concat([s1,s2], axis=1)
```

disini kita menggabungkan data kumpulan judul dan data teks dari setiap artikel yang telah diperoleh kedalam suatu data frame yang nantinya akan diexport menjadi csv

```python
crawling_list
```

```python
crawling_list.to_csv('crawling.csv')
```

3.	Setelah itu, gunakan scrapy untuk melakukan crawling website https://store.playstation.com/en-id 

```python
import scrapy

class ps_deals(scrapy.Spider):
    name = "deals"
    start_urls = ['https://store.playstation.com/en-id/category/805ff1ce-4e9f-4b73-8508-7cd344c40afb/']

    def parse(self, response):
        num_page = response.css('span.psw-fill-x::text')[-1].get()
        num_page = int(num_page)

        url = response.url

        for offset in range(1,num_page+1):
            yield scrapy.Request(url = url+str(offset), callback = self.parse_details)


    def parse_details(self, response):
        for post in response.css('.psw-product-tile__details'):
            yield {
                'title' : post.css('.psw-t-body::text').get(),
                'deal_price' : post.css('.psw-m-r-3::text').get(),
                }
```

Berikut adalah salah satu dari isi folder spider. berfungsi untuk mendapatkan data nama semua games yang ada pada website playstation dan harga diskon yang diberikan. Untuk dapat menjalankan data ini, kita pertama installing scrapy pada environment kita. Lalu kita melakukan crawling melalui terminal yang ada. Saya menggunakan pycharm untuk memudahkan dalam menyeting environment dan installing scrapy sendiri. Dalam terminal kita memasukkan 'scrapy crawl <name>'
